```python
#Join (Here notification is only received after uploading is complete)
from threading import Thread
from time import sleep

def upload():
    print('Uploading starts')
    sleep(3)
    print('video uploaded')
    
def notification():
    print('Sending notifications to subscribers code')
    
t1=Thread(target=upload)
t2=Thread(target=notification)
t1.start()
t1.join()
t2.start()
t2.join()

#Main thread will wait for t2 to complete which will wait for t1 to complete
for i in range(4):
    print('Hello')
```

```python
#Why is multithreading efficient? 
from threading import Thread
import time

def square(num):
    print('finding square..')
    time.sleep(1)
    print(num**2)

def cube(num):
    print('finding cube..')
    time.sleep(1)
    print(num**3)
begin=time.time()  
t1=Thread(target=square,args=(3,))
t2=Thread(target=cube, args=(3,))
t1.start()
t2.start()
t1.join()
t2.join()
print('Total time taken: ', time.time()-begin)
```

```python
#Solving the Race Condition
from threading import *
from time import sleep
lock=Lock()
class Bus:
    def __init__(self,name,available_seats,l):
        self.available_seats=available_seats
        self.name=name
        self.l=l
    def reserve(self,need_seats):
        self.l.acquire()
        print('Available Seats are:', self.available_seats)
        if self.available_seats>=need_seats:
            nm=current_thread().name
            #Ticket booking code goes here. Update in the database
            print(f"{need_seats} are allocated to {nm}")
            self.available_seats-=need_seats
        else:
            print('Sorry seats are not available')
        self.l.release()
b1=Bus('Mahalaxmi travels',2,lock)
t1=Thread(target=b1.reserve,args=(2,),name='Jay')
t2=Thread(target=b1.reserve, args=(1,),name='Raj')
t1.start()
t2.start()
```

```python
#Solving the Race Condition with RLock
from threading import *
from time import sleep
rl=RLock()
class Bus:
    def __init__(self,name,available_seats,rlock):
        self.available_seats=available_seats
        self.name=name
        self.rlock=rlock
    def reserve(self,need_seats):
        self.rlock.acquire()
        self.rlock.acquire()
        print('Available Seats are:', self.available_seats)
        if self.available_seats>=need_seats:
            nm=current_thread().name
            #Ticket booking code goes here. Update in the database
            print(f"{need_seats} are allocated to {nm}")
            self.available_seats-=need_seats
        else:
            print('Sorry seats are not available')
        self.rlock.release()
        self.rlock.release()
b1=Bus('Mahalaxmi travels',2,rl)
t1=Thread(target=b1.reserve,args=(2,),name='Jay')
t2=Thread(target=b1.reserve, args=(1,),name='Raj')
t1.start()
t2.start()
```

```python
#Need for RLock 
from threading import * 
from time import sleep
l=RLock()
def get_first_line():
    l.acquire()
    print('First')
    l.release()

def get_second_line():
    l.acquire()
    print('Second')
    l.release()

def main():
    l.acquire()
    get_first_line()
    get_second_line()
    l.release()
t1=Thread(target=main)
t1.start()
```

```python
#Thread Communication (Event Class)
import threading 
import time
e=threading.Event() #Step-1

def task():
    print("Game is started")
    time.sleep(3)
    e.set()

def end():
    e.wait() #Step 2
    if e.is_set(): #Step-3
        print('Game has ended')

t1=threading.Thread(target=task)
t2=threading.Thread(target=end)

t1.start()
t2.start()
```

```python
#Thread Communication  (Queue Condition)
from time import sleep
from threading import Thread
from queue import Queue

def producer(my_que):
    print('producer: running')
    n=int(input('Enter number of students: '))
    for i in range(n):
        marks=float(input('Enter marks'))
        my_que.put(marks)
    my_que.put(None)
    
    print('producer: end')

def consumer(my_que):
    print('consumer: running')
    while True:
        item=my_que.get()
        if item is None: 
            break
        print(f"Got {item}")
    print('consumer: end')
my_que=Queue()
t1=Thread(target=producer, args=(my_que,))
t2=Thread(target=consumer, args=(my_que,))
t1.start()
t2.start()
```
